```python
from asr import ASR
import mne
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#LOADING DATA
#---------------
#Get data path and load data
data_path = os.getcwd()+'/Data/'
data_file = 'test.bdf'
mat_fname = os.path.join(data_path, data_file)
raw = mne.io.read_raw_bdf(mat_fname)
raw.load_data()

#drop channels not in Unicorn
ch_names = ['Fz', 'C3', 'Cz', 'C4', 'Pz', 'PO7', 'Oz', 'PO8']
raw.drop_channels(np.setdiff1d(raw.info.ch_names,ch_names))
raw.info
```

```python
#plot raw data
plt.figure(figsize=(20,5))
plt.plot(raw.get_data().T);
```

```python
#PREPROCESSING
#---------------
```

```python
#apply bandpass filter
raw.filter(0.5, 100, fir_design='firwin')
```

```python
#plot filtered data
plt.figure(figsize=(20,5))
plt.plot(raw.get_data().T);
```

```python
#initialize Artifact Subspace Reconstruction
asr = ASR()
```

```python
#plot calibration data
raw_calib = raw.copy()
raw_calib.crop(0, 60.)
x_c = asr.clean_windows(raw_calib)
plt.figure(figsize=(20,5))
plt.plot(x_c.T);
```

```python
# clean up signal and plot - reconstruct parts with high amplitude noise such as eyeblinks
asr.calibrate()
out = asr.clean(raw)
plt.figure(figsize=(20,5))
plt.plot(out.T);
```
